```python
import hashlib

def calculate_sha256(input_string):
    # 1. Convert the string to bytes
    byte_input = input_string.encode()
    
    # 2. Create the SHA-256 object and hash the data
    hash_object = hashlib.sha256(byte_input)
    
    # 3. Get the hexadecimal representation
    hex_dig = hash_object.hexdigest()
    
    print(f"Input: {input_string}")
    print(f"Hash:  {hex_dig}\n")

# Testing the Avalanche Effect
calculate_sha256("Hello World")
calculate_sha256("hello world") # Only the 'H' changed, but the hash is totally different
```
