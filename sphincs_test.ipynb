```python
import os
import sys
# import pyspx.shake_128f as sphincs
# import pyspx.shake_192f as sphincs
import pyspx.shake_256f as sphincs

def demo_sphincs_signature():
    print("====================================================")
    print("        SPHINCS+ Post-Quantum Signature Demo        ")
    print("====================================================\n")

    # 1. Define data payload to be signed
    message = b"fear not, for I am with you; be not dismayed, for I am your God; I will strengthen you, I will help you, I will uphold you with my righteous right hand."
    print(f"[+] Original Message: '{message.decode()}'")

    # 2. Key Generation with explicit Seed
    print("\n[*] Generating SPHINCS+ Keypair...")
    
    # SPHINCS+ requires a physical seed to initialize its random tree layers
    # seed = os.urandom(48)
    # seed = os.urandom(72) 
    seed = os.urandom(96) 
    public_key, secret_key = sphincs.generate_keypair(seed)

    print(f"    -> Public Key Size : {len(public_key)} bytes")
    print(f"    -> Secret Key Size : {len(secret_key)} bytes")
    print(f"    -> Public Key (Hex): {public_key[:16].hex()}...")

    # 3. Create Digital Signature
    print("\n[*] Signing the message...")
    signature = sphincs.sign(message, secret_key)
    
    print(f"    -> Signature Size: {len(signature)} bytes")
    print(f"    -> Signature (Hex): {signature[:32].hex()}...")

    # 4. Verification Pathway (Success Case)
    print("\n[*] Verifying signature with valid public key...")
    is_valid = sphincs.verify(message, signature, public_key)
    print(f"    -> Verification Result: {'SUCCESS ✅' if is_valid else 'FAILED ❌'}")

if __name__ == "__main__":
    demo_sphincs_signature()
```
