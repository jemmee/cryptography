```python
import os
import binascii

# Generate 32-byte (256-bit) key for AES-256
key = os.urandom(32)
print("Key (hex):", binascii.hexlify(key).decode())

# For symmetric encryption (e.g., Fernet)
from cryptography.fernet import Fernet
fernet_key = Fernet.generate_key()  # Internally uses os.urandom
print("Fernet key:", fernet_key.decode())
```

```python
import secrets

# 32-byte key
key = secrets.token_bytes(32)
print("Key (hex):", key.hex())

# URL-safe token (e.g., for passwords/salts)
token = secrets.token_urlsafe(32)
print("URL-safe token:", token)
```
