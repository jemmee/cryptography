```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
import os

# 1. SETUP
# For AES-256-XTS, we need a 512-bit key (64 bytes)
key = os.urandom(64)
# The "Tweak" (16 bytes) is usually the Sector ID or Block Address
tweak = os.urandom(16) 

data = b"The kingdom of heaven is like a mustard seed, which a man took and planted in his field." # Must be at least 16 bytes

# 2. ENCRYPTION
cipher = Cipher(algorithms.AES(key), modes.XTS(tweak))
encryptor = cipher.encryptor()
ciphertext = encryptor.update(data) + encryptor.finalize()

print(f"Tweak (hex):      {tweak.hex()}")
print(f"Ciphertext (hex): {ciphertext.hex()}")

# 3. DECRYPTION
decryptor = cipher.decryptor()
plaintext = decryptor.update(ciphertext) + decryptor.finalize()

print(f"Decrypted:        {plaintext.decode()}")
```
