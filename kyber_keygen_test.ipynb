```python
import jpype
import jpype.imports
import os

# Use absolute path to ensure the JAR is found
jar_path = os.path.abspath('./bcprov-jdk18on-1.83.jar')
if not jpype.isJVMStarted():
    jpype.startJVM(classpath=[jar_path])

from org.bouncycastle.pqc.crypto.mlkem import MLKEMKeyGenerationParameters, MLKEMKeyPairGenerator, MLKEMParameters
from java.security import SecureRandom

# Available options in v1.83:
# MLKEMParameters.ml_kem_512
# MLKEMParameters.ml_kem_768
# MLKEMParameters.ml_kem_1024

random = SecureRandom()
params = MLKEMKeyGenerationParameters(random, MLKEMParameters.ml_kem_768)

kpg = MLKEMKeyPairGenerator()
kpg.init(params)
keyPair = kpg.generateKeyPair()

print("ML-KEM-768 Keypair successfully generated!")
```
