```python
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives import hashes

def main():
    # 1. Generate an RSA Private/Public Key Pair
    print("Generating 2048-bit RSA key pair...")
    private_key = rsa.generate_private_key(
        public_exponent=65537,
        key_size=2048
    )
    public_key = private_key.public_key()
    print("Keys generated successfully.\n")

    # 2. Define the payload
    secret_message = b"Jesus wept"
    print("Original Plaintext: " + secret_message.decode())

    # 3. Encrypt using RSA + OAEP Padding
    # We configure OAEP with MGF1 (Mask Generation Function) and SHA-256
    ciphertext = public_key.encrypt(
        secret_message,
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None # Optional shared application context data
        )
    )
    
    print("\n--- Encrypted Ciphertext (Hex) ---")
    print(ciphertext.hex())
    print(f"Ciphertext Length: {len(ciphertext)} bytes")

    # 4. Decrypt using the Private Key + identical OAEP parameters
    try:
        decrypted_message = private_key.decrypt(
            ciphertext,
            padding.OAEP(
                mgf=padding.MGF1(algorithm=hashes.SHA256()),
                algorithm=hashes.SHA256(),
                label=None
            )
        )
        print("\n--- Decryption Successful ---")
        print("Decrypted Plaintext: " + decrypted_message.decode())
        
    except Exception as e:
        print(f"\nDecryption failed: {e}")

if __name__ == "__main__":
    main()
```
