```python
import hashlib

def demo_shake512_interactive():
    print("--- SHAKE-512 (Extendable Output) ---")
    
    # 1. Get the data to hash
    user_text = input("Enter text to hash: ")
    
    # 2. Get the desired output length
    try:
        bit_length = int(input("How many BITS of output do you want? (e.g., 512, 1024, 2048): "))
        byte_length = bit_length // 8
    except ValueError:
        print("Invalid number. Defaulting to 512 bits.")
        byte_length = 64

    # 3. Create the hash
    shake = hashlib.shake_256(user_text.encode('utf-8'))
    
    # 4. Squeeze out the desired amount of data
    # Unlike SHA3-512, SHAKE can 'squeeze' infinitely
    hex_result = shake.hexdigest(byte_length)

    print(f"\n[SHAKE-256 result for {bit_length} bits]:")
    print(hex_result)
    print(f"\nResulting hex string length: {len(hex_result)} characters")

if __name__ == "__main__":
    demo_shake512_interactive()
```
