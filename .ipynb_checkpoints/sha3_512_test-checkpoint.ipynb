```python
import hashlib

def sha3_512():
    print("--- SHA3-512 Hash Generator ---")
    
    # Using standard input() so the text is visible as you type
    user_text = input("Enter the text you want to hash: ")
    
    if not user_text:
        print("Error: You didn't enter any text!")
        return

    # 1. Convert the string to bytes (UTF-8)
    data_bytes = user_text.encode('utf-8')
    
    # 2. Generate the SHA3-512 hash
    # SHA3-512 is the strongest version of the SHA-3 family
    sha3_hash = hashlib.sha3_512(data_bytes).hexdigest()

    # 3. Output the results
    print("\n" + "="*30)
    print(f"Original Text: {user_text}")
    print(f"Hash Length:   {len(sha3_hash)} characters")
    print(f"SHA3-512 Hash:\n{sha3_hash}")
    print("="*30)

if __name__ == "__main__":
    sha3_512()
```
