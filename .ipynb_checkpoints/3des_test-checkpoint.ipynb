```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
import os

def demo_3des():
    # 1. Setup: 3DES requires a 24-byte key (3 * 8 bytes) and an 8-byte IV
    # In a real app, use a secure key; here we generate random ones.
    key = os.urandom(24) 
    iv = os.urandom(8)   
    
    # 2. The Message (Must be a multiple of 8 bytes for 3DES)
    plaintext = b"SECRET12" # 8 bytes
    print(f"Original Text: {plaintext.decode()}")

    # 3. Encryption
    cipher = Cipher(algorithms.TripleDES(key), modes.CBC(iv), backend=default_backend())
    encryptor = cipher.encryptor()
    ciphertext = encryptor.update(plaintext) + encryptor.finalize()
    
    print(f"Ciphertext (Hex): {ciphertext.hex()}")

    # 4. Decryption
    decryptor = cipher.decryptor()
    decrypted_text = decryptor.update(ciphertext) + decryptor.finalize()
    
    print(f"Decrypted Text: {decrypted_text.decode()}")

if __name__ == "__main__":
    demo_3des()
```
