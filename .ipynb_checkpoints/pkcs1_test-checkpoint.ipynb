```python
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.asymmetric import padding
from cryptography.hazmat.primitives import serialization

# 1. LOAD YOUR PRIVATE KEY
with open("private_key.pem", "rb") as key_file:
    private_key = serialization.load_pem_private_key(
        key_file.read(),
        password=None,
    )

message = b"This is the sign of the covenant I have established between me and all life on the earth."

# 2. SIGN using PKCS#1 v1.5 padding
signature = private_key.sign(
    message,
    padding.PKCS1v15(),  # This is the PKCS#1 v1.5 standard
    hashes.SHA256()
)

print(f"Signature (hex): {signature.hex()[:50]}...")

# 3. VERIFY using the Public Key
public_key = private_key.public_key()

try:
    public_key.verify(
        signature,
        message,
        padding.PKCS1v15(),
        hashes.SHA256()
    )
    print("Verification Successful!")
except Exception as e:
    print(f"Verification Failed: {e}")
```
