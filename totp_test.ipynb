```python
import pyotp
import time
import qrcode

# 1. Generate a random base32 secret (compatible with Google Authenticator)
# In a real app, you would save this secret in your database for this specific user.
secret = pyotp.random_base32()
print(f"Your Secret Key: {secret}")

# 2. Setup the TOTP object
totp = pyotp.TOTP(secret)

# 3. Create a Provisioning URI (to scan with an app)
# name: the user's email or username
# issuer_name: your company or app name
uri = totp.provisioning_uri(name="user@example.com", issuer_name="MySecureApp")
print(f"Provisioning URI: {uri}")

# (Optional) Generate a QR Code image for the user to scan
img = qrcode.make(uri)
img.save("totp_test.png")
print("QR Code saved as 'totp_qr.png'. Scan it with Google Authenticator or Authy.")

# 4. Simulation: Generate and Verify
current_otp = totp.now()
print(f"\nThe current OTP generated by your script is: {current_otp}")

# 5. Verification test
# In a real login form, you'd get this from input()
user_input = input("Enter the 6-digit code from your app: ")

if totp.verify(user_input, valid_window=1):
    print("✅ Success: Code is valid!")
else:
    print("❌ Failed: Invalid or expired code.")

# 6. Demonstration of Expiration
print("\nWaiting 30 seconds for the code to expire...")
time.sleep(31)
if not totp.verify(current_otp):
    print("✅ Test Passed: The old code is no longer valid.")
```
