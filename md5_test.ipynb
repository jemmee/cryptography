```python
# md5_demo.py
# Simple demonstration of MD5 hashing in Python using the built-in hashlib library

import hashlib

def md5_hash(text: str) -> str:
    """Compute the MD5 hash of a given string and return it as a hexadecimal string."""
    # Encode the string to bytes (required by hashlib)
    encoded_text = text.encode('utf-8')
    
    # Create an MD5 hash object
    md5 = hashlib.md5()
    
    # Update the hash object with the bytes
    md5.update(encoded_text)
    
    # Get the hexadecimal representation of the hash
    return md5.hexdigest()

# Interactive demo
if __name__ == "__main__":
    print("MD5 Hash Demo\n")
    
    # Default example
    default_text = "Hello, world!"
    print(f"Example input: '{default_text}'")
    print(f"MD5 hash: {md5_hash(default_text)}\n")
    
    # Ask user for input
    user_input = input("Enter your own text to hash (or press Enter to skip): ").strip()
    
    if user_input:
        print(f"\nYour input: '{user_input}'")
        print(f"MD5 hash: {md5_hash(user_input)}")
    else:
        print("\nSkipping custom input.")
```
