```python
import hashlib

def hash_val(val):
    return hashlib.sha256(val.encode()).hexdigest()

# 1. Setup
secret_seed = "He that dwelleth in the secret place of the Most High Shall abide under the shadow of the Almighty."
chain_length = 5
chain = [secret_seed]

# Generate the chain forward
for i in range(chain_length):
    chain.append(hash_val(chain[i]))

# The server only knows the last value
server_stored_hash = chain.pop() 
print(f"Server starts with: {server_stored_hash[:10]}...")

# 2. Login simulation (using the chain backwards)
while chain:
    user_submission = chain.pop()
    print(f"\nUser submits: {user_submission[:10]}...")
    
    # Server verifies: H(submitted) == stored
    if hash_val(user_submission) == server_stored_hash:
        print("AUTHENTICATED!")
        server_stored_hash = user_submission # Update stored hash
    else:
        print("REJECTED!")
```
