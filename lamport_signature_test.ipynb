```python
import hashlib
import secrets

def h(val):
    return hashlib.sha256(val).hexdigest()

# 1. KEY GEN (For a single bit of data)
# We generate two secrets (Private Key)
secret_0 = secrets.token_bytes(32)
secret_1 = secrets.token_bytes(32)

# We hash them (Public Key)
pub_0 = h(secret_0)
pub_1 = h(secret_1)

print("--- Lamport One-Time Signature ---")
print(f"Public Key Part 0: {pub_0[:16]}...")
print(f"Public Key Part 1: {pub_1[:16]}...")

# 2. SIGNING
# We want to sign the bit '1' (representing an 'Attack' order or a 'Census' entry)
message_bit = 1
signature = secret_1 if message_bit == 1 else secret_0

print(f"\nSigning bit {message_bit}...")
print(f"Signature (Revealed Secret): {signature.hex()[:16]}...")

# 3. VERIFICATION
# The receiver hashes the signature and compares it to the Public Key
verification_hash = h(signature)
if message_bit == 1 and verification_hash == pub_0: # Wrong bit check
    print("VERIFICATION FAILED")
elif message_bit == 1 and verification_hash == pub_1:
    print("VERIFICATION SUCCESS: The bit is authentic.")
```
