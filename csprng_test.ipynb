```python
import secrets
import binascii
import os

def csprng_showcase():
    # 1. Generating a Secure Random Integer
    # Range: 0 to 1,000,000. Use this for game logic or IDs.
    secure_int = secrets.randbelow(1000000)
    print(f"Secure Random Integer: {secure_int}")

    # 2. Generating a Secure Token (Hexadecimal)
    # Use this for password reset tokens or session IDs.
    secure_token = secrets.token_hex(16) 
    print(f"Secure Hex Token: {secure_token}")

    # 3. Generating Raw Entropy (Bytes)
    # Use this for cryptographic keys or salts.
    raw_bytes = secrets.token_bytes(32)
    print(f"Raw Entropy (Hex): {binascii.hexlify(raw_bytes).decode()}")

    # 4. The "Under the Hood" source
    # This pulls directly from the OS Kernel (Netfilter/Hardware noise)
    os_random = os.urandom(16)
    print(f"Direct OS Entropy: {binascii.hexlify(os_random).decode()}")

if __name__ == "__main__":
    print("--- Cryptographically Secure PRNG Demo ---")
    csprng_showcase()
```
